```python
import numpy as np
from collections import Counter
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# generate dataset
X,y=make_blobs(n_samples=150,n_features=2,cluster_std=2,centers=2,random_state=42)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
def euclidean_distance(p1,p2):
    return np.sqrt(np.sum((np.array(p1) - np.array(p2))**2))
```

```python
def knn(training_data,training_labels,test_point,k):
    distances=[]

    for i in range(len(training_data)):
        dist=euclidean_distance(test_point,training_data[i])
        distances.append((dist,training_labels[i]))

    # we want to sort by distances
    # since tuple consist of dist and label we will use key dist which is x[0]
    distances.sort(key=lambda x:x[0])

    # _ for distance and label for lables
    # we want to store only labels
    k_nearest_labels=[label for _,label in distances[:k]]
    # most common 1 element
    label, count = Counter(k_nearest_labels).most_common(1)[0]
    return label, count
```

```python
data_point=[2,6]
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(data_point[0],data_point[1])
plt.show()
```

```python
k = 3
label,count=knn(X,y,data_point,k)
print(label)
print(count)
```
